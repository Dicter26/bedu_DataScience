```python
import time

object_date = {}

endpoint = 'https://api.nasa.gov/neo/rest/v1/feed?'
payload = {'start_date': '1995-01-01',
           'end_date': '1995-01-07',
           'api_key': 'cjjlrLngF0fSCT2fsCJnSKHrBtohseZUWfd1osYt'}

for i in range(0, 9):
  try:
    r = requests.get(endpoint, params= payload)
    if r.status_code == 200:
      json = r.json()
      data = json['near_earth_objects']
      print(json['links']['next'])
      object_date[i] = data
      endpoint = json['links']['next']
  except:
    continue
```
